```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('winequality-red.csv')
```

```python
dataset.info()
```

```python
df = pd.DataFrame(dataset)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df['fixed acidity'] = sc.fit_transform(df[['fixed acidity']])
df['volatile acidity'] = sc.fit_transform(df[['volatile acidity']])
df['citric acid'] = sc.fit_transform(df[['citric acid']])
df['residual sugar'] = sc.fit_transform(df[['residual sugar']])
df['chlorides'] = sc.fit_transform(df[['chlorides']])
df['free sulfur dioxide'] = sc.fit_transform(df[['free sulfur dioxide']])
df['total sulfur dioxide'] = sc.fit_transform(df[['total sulfur dioxide']])
df['density'] = sc.fit_transform(df[['density']])
df['pH'] = sc.fit_transform(df[['pH']])
df['sulphates'] = sc.fit_transform(df[['sulphates']])
df['alcohol'] = sc.fit_transform(df[['alcohol']])
```

```python
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X[0]
```

```python
y[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
X_test[0]
```

```python
y_test[0]
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=0)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
